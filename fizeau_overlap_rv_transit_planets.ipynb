```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib qt
```

```python
data_file_name = "./data/nasa_exoplanet_archive_20220308.csv"
```

```python
df = pd.read_csv(data_file_name, header = 34)
```

```python
df["st_spectype"]
```

```python
df.keys()
```

```python
df["angles"] = np.divide(df["pl_orbsmax"],df["sy_dist"])
```

```python
idx_rv = df["discoverymethod"] == "Radial Velocity"
idx_transit = df["discoverymethod"] == "Transit"
```

```python
df["discoverymethod"].value_counts()
```

```python
rv_planets_only = df[idx_rv]
rv_planets_only["Angle (arcsec)"] = rv_planets_only["angles"]
rv_planets_only["DEC (degrees)"] = rv_planets_only["dec"]
rv_planets_only["DEC (degrees)"] = rv_planets_only["dec"]
rv_planets_only["Msin(i)"] = rv_planets_only["pl_bmassj"]
rv_planets_only["Ks mag"] = rv_planets_only["sy_kmag"]
```

```python
idx_rv_bright = rv_planets_only["Ks mag"] < 4.5
rv_planets_bright = rv_planets_only[idx_rv_bright]
```

```python
plt.hist(df[idx_rv]["angles"],bins=300)
#plt.hist(df[idx_transit]["angles"],bins=200,color="r", alpha=0.5)
plt.axvline(x=0.037, linestyle=":", color="k")
plt.axvline(x=0.101, linestyle=":", color="k")
plt.xlim([0.0,0.3])
plt.ylabel("Count")
plt.xlabel("Widest angle (arcsec)")
plt.show()
```

```python
plt.clf()

sns.set_theme(style="whitegrid")

# Load the example planets dataset
planets = rv_planets_only
#planets = rv_planets_bright

#cmap = sns.color_palette("rocket", as_cmap=True)
palette = sns.color_palette("rocket")
cmap = palette.reverse()
g = sns.relplot(
    data=planets,
    x="Angle (arcsec)", y="DEC (degrees)",
    hue="Ks mag", size="Msin(i)",
    palette=cmap, sizes=(10, 200), edgecolor='k', height=4, aspect=1.5
)

g.set(xscale="log")
#plt.axvline(x=0.037, color="k", linewidth = 7, alpha=0.5)
plt.axvline(x=0.101, color="k", linewidth = 7, alpha=0.5)
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
plt.xlim([1e-4,10.])
plt.ylim([-90,90])
#g.set(rc = {'figure.figsize':(5,8)})

g.savefig("junk_all_w_1_lines.pdf")
```
